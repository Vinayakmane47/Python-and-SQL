```python
##=======================================FUNCTION DEFINATION=============================================================##
 
# Adding new word : 
def new_word() : 
    f = open('words.txt','r')
    data = json.load(f)
    add_word = input('Enter a word here : ')
    add_meaning = input('Enter meaning of word here : ')
    if add_word in data.keys() : 
        print('Word already exists in dictionary ')
    else : 
        data[add_word] = add_meaning 
    f = open('words.txt', 'w')
    json.dump(data,f)
    f.close()

#----------------------------------------------------------------------------------------------------------------------------

# Find the meaning of word : 
def word_mean() : 
    f = open('words.txt','r')
    meaning = json.load(f)
    letter = input('Enter a word here : ')
    for i in meaning.keys() : 
        if letter == i : 
            print('Meaning of',i,'is : ', meaning[i])
            break
    else : 
        print('Word is not found in dictionary')
    f.close()


#---------------------------------------------------------------------------------------------------------------------------
# Update a meaning of Word 

def update(word) : 
    f = open('words.txt','r')
    update_mean = json.load(f)
    meaning = input('Enter meaning of word : ')
    update_mean[word] = meaning
    f = open('words.txt','w')
    json.dump(update_mean,f)
    f.close()

#-------------------------------------------------------------------------------------------------------------------------
def menu() : 
    print('[0] Start Application')
    print('[1] Add a New Word')
    print('[2] Find a Meaning ')
    print('[3] Update a Word ')
    print('[4] Exit the Programme')
    



## ======================================CREATING words.txt FILE IF NOT EXISTS==============================================    

import json
import os.path
from os import path

# Checking whether the file is exists or not , and if do exists checking if it is empty or not.  

if path.exists('words.txt') is True :
    if os.stat('words.txt').st_size == 0:
        f = open('words.txt','w')
        data1 = dict()
        json.dump(data1,f)
        
else : 
    f = open('words.txt','w')
    data1 = dict()
    json.dump(data1,f)
    
##=======================================STARTING APPLICATION================================================================#
print('Entering into main menu : ')
menu()
option = int(input('Enter option : '))
if option == 0 : 
    print('Application is Running Now : ')
## ===================================PROGRAMME EXECUTION============================================================##
while option < 4 :
    print('*'*120)
    print('Entering into menu option ')
    menu()
    option = int(input('Enter option : '))
    while True : 
        if option == 1 : 
            new_word()
            print('New Word is added Succesfully in dictionary')
            break
        elif option == 2 : 
            word_mean()
            print('Option 2 ran Succesfully')
            break
        elif option == 3 :
            word = input('Enter a word to update :')
            update(word)
            print('Existing  Word is updated Succesfully in dictionary')
            break
        elif option == 0  : 
            print('Application is Running ')
            break
        elif option == 4 : 
            print('Application is closed ')
            break



      
```
