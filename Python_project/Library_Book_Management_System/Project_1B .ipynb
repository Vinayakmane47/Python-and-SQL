```python


#===================================IMPORTING MODULES=====================================================
from tkinter import * 
from tkinter import ttk
import sqlite3
import tkinter.messagebox as tkmsg
#====================================FUNCTION DEFINATION================================================
def database(): 
    global cur, con
    con = sqlite3.connect('library.db')
    cur = con.cursor()
    cur.execute("CREATE TABLE IF NOT EXISTS book (Booktitle TEXT , Author TEXT , Year INT  , ISBN INT )")
    con.commit()
   
    
def add(): 
    
    if BOOKTITLE.get()=="" or AUTHOR.get() == "" or YEAR.get() =="" or ISBN.get() =="" :
        text_results.config(text="Please enter all the fields", fg = "red")
    else : 
        database()
        cur.execute("INSERT INTO book (Booktitle , Author , Year , ISBN )  VALUES (?,?,?,?)", (str(BOOKTITLE.get()), str(AUTHOR.get()), str(YEAR.get()), str(ISBN.get())))
        con.commit()
        BOOKTITLE.set("")
        AUTHOR.set("")
        YEAR.set("")
        ISBN.set("")
        cur.close()
        text_results.config(text = "Record inserted !" , fg = "green")
        
def display() :
    database()
    tree.delete(*tree.get_children()) 
    cur.execute("SELECT * FROM book ORDER BY BOOKTITLE ASC ")
    fetch = cur.fetchall()
    
    for data in fetch : 
        tree.insert('','end', values = (data[0], data[1], data[2], data[3]))
    cur.close()
    text_results.config(text = "Data succesfully fetched !" , fg = "green")
    
    
def Exit() : 
    message = tkmsg.askquestion("Do you really want to close application ?  (y/n)", icon = "warning")
    if message == "yes" : 
        root.destroy()

def issue() : 
    item = tree.selection()[0]
    tree.delete(item)
    database()
    cur.execute("DELETE FROM book WHERE Booktitle = ?", (item,))
    text_results.config(text = "Book is issued Succesfully !" , fg ="green")
        

        
def search() : 
    database()
    records = BOOKTITLE.get()
    cur.execute("SELECT * FROM book WHERE Booktitle = ? ", (records,))
    fetch = cur.fetchall()
    if records == "" : 
        text_results.config(text = "Please Enter Booktitle" , fg = "red")
        
    else : 
        for data in fetch :
            tree.delete(*tree.get_children())
            tree.insert("","end" , values = (data[0], data[1], data[2] , data[3]))
        BOOKTITLE.set("")
        cur.close()
        text_results.config(text = "Searching Completed" , fg = "green")


#==================================CREATING WINDOW ======================================================

root = Tk()
root.geometry('1000x600')
root.maxsize(1000,600)
root.minsize(1000,600)
#===================================FRAMES==============================================================
Top = Frame(root, height = 50, width = 800 ,bg = 'Blue' ,bd = 8 , relief = SUNKEN )
Top.pack(side = TOP , anchor = N,  fill = X , pady = 20)
left = Frame(root,height =500 , width =600, bg = 'Red', bd = 8 , relief = SUNKEN)
left.pack(side = LEFT , padx = 20)
Right = Frame(root, width = 600 ,bd = 8 , relief = SUNKEN)
Right.pack(side =RIGHT, padx = 20, fill =Y)
forms = Frame(left, width=300, height = 450 , bd = 8 , relief = SUNKEN)
forms.pack(side = TOP)
buttons = Frame(left, width = 600 , height = 100 ,  bd = 8 , bg = 'Gray', relief = SUNKEN )
buttons.pack(side = BOTTOM)



# ==========================================VARIABLES=================================================== 
BOOKTITLE = StringVar()
AUTHOR = StringVar()
YEAR   = StringVar()
ISBN   = StringVar()


#============================================LABELS======================================================

label1 = Label(Top, text = 'Library Book Management System', bg = 'Red', fg = 'White', font = 'Arial 25')
label1.pack(fill = X, pady = 5 , padx = 5)

l1 = Label(forms, text = 'Booktitle : ' , font = 'Arial 16' , bd = 15 )
l1.grid(row = 0 , column = 0 , padx = 5, pady = 10)
l2 = Label(forms, text = 'Author : ' , font = 'Arial 16' , bd = 15  )
l2.grid(row = 1 , column = 0 ,padx = 5, pady = 10)
l3 = Label(forms, text = 'Year : ' , font = 'Arial 16' , bd = 15 )
l3.grid(row = 2 , column = 0,padx = 5, pady = 10)
l4 = Label(forms, text = 'ISBN : ' , font = 'Arial 16' , bd = 15  )
l4.grid(row = 3 , column = 0, padx = 5, pady = 10)
text_results = Label(left)
text_results.pack(side = TOP)

# ========================================ENTRY WIDGETS=================================================

e1 = Entry(forms, width = 50, textvariable = BOOKTITLE).grid(row = 0 , column = 1,padx = 5, pady = 10)
e2 = Entry(forms,width = 50 , textvariable = AUTHOR).grid(row = 1 , column = 1,padx = 5, pady = 10)
e3 = Entry(forms,width = 50 , textvariable = YEAR).grid(row = 2 , column = 1,padx = 5, pady = 10)
e4 = Entry(forms,width = 50, textvariable = ISBN).grid(row = 3 , column = 1,padx = 5, pady = 10)


#===================================BUTTON WIDGETS======================================================
b1 = Button(buttons,text = 'Display ' , width = 10 ,  command = display)
b1.grid(row = 0 , column = 1, padx = 5 )

b2 = Button(buttons,text = 'Search ' , width = 10 , command = search)
b2.grid(row = 0 , column = 2 ,padx = 5)

b3 = Button(buttons,text = ' Add ' , width = 10 ,  command = add)
b3.grid(row = 0 , column = 3 ,padx = 5)

b4 = Button(buttons,text = 'Issue ' , width = 10 , command = issue)
b4.grid(row = 0 , column = 4,padx = 5)

b4 = Button(buttons,text = 'Exit ' , width = 10 , command = Exit)
b4.grid(row = 0 , column = 5 ,padx = 5)



# =========================================TREEVIEW====================================================
tree = ttk.Treeview(Right , columns=("Booktitle" , "Author", "Year", "ISBN"))
scrollbar = Scrollbar(Right, orient = VERTICAL)
scrollbar.pack(side = RIGHT , fill = Y)
scrollbar.config(command = tree.yview)
tree.heading("Booktitle", text = "Booktitle", anchor = W)
tree.heading("Author" , text = "Author", anchor = W)
tree.heading("Year", text = "Year", anchor = W)
tree.heading("ISBN", text = "ISBN" , anchor =W)

tree.column('#0', stretch = NO , minwidth = 0 , width = 0)
tree.column('#1', stretch = NO , minwidth = 0 , width = 80 , anchor = W)
tree.column('#2', stretch = NO , minwidth = 0 , width = 100 , anchor = W)
tree.column('#3', stretch = NO , minwidth = 0 , width = 120 , anchor = W)
tree.column('#4', stretch = NO , minwidth = 0 , width = 120  , anchor = W)
tree.pack(side = RIGHT,fill = Y)

#========================================INVOKING TKINTER================================================
root.mainloop()
```
