### Q2. Checking and Cleaning Data Quality

#### A)  Null values : 

```python

# Summary of Null spaces present in Customer  details file  : 
df1_count_null = pd.DataFrame([['customer_id', df1['customer_id'].isnull().sum()] , \
                ['Gender',df1['Gender'].isnull().sum()] ,['age',df1['age'].isnull().sum()] , \
                ['driving_licence_present',df1['driving_licence_present'].isnull().sum()] ,  \
                 ['region_code', df1['region_code'].isnull().sum()] , \
               ['previously_insured', df1['previously_insured'].isnull().sum()] , \
                ['vehicle_age', df1['vehicle_age'].isnull().sum()] , \
                ['vehicle_damage', df1['vehicle_damage'].isnull().sum()]] , columns = ['Column Name' , 'Null Spaces'])

df1_count_null
```

```python
# Summary of Null spaces present in Customer details file  : 
df2_count_null = pd.DataFrame([['customer_id', df2['customer_id'].isnull().sum()] , \
                ['annual_premium',df2['annual_premium'].isnull().sum()] , \
                ['sales_channel_code',df2['sales_channel_code'].isnull().sum()] ,  \
                 ['vintage', df2['vintage'].isnull().sum()] , \
               ['response', df2['response'].isnull().sum()]] , columns = ['Column Name' , 'Null Spaces'])

df2_count_null
```

```python
# Dropping null values from customer_id column from first dataframe 
df1 = df1.dropna(subset= ['customer_id'])
df1
```

```python
# Dropping null values from customer_id column from second dataframe 
df2 = df2.dropna(subset= ['customer_id'])
df2
```

```python
# Replacing null values in numeric  column by mean in customer details file  : 
df1['age'] = df1['age'].fillna(df1['age'].mean())
df2['annual_premium'] = df2['annual_premium'].fillna(df2['annual_premium'].mean())
df2['vintage'] = df2['vintage'].fillna(df2['vintage'].mean())


```

```python
#  Replacing null values in categorial values  by mode in customer details file    : 
df1 = df1.fillna(df1.mode().iloc[0])
df2 = df2.fillna(df2.mode().iloc[0])

```

#### B) Outliers

```python
df1.describe()
```

```python
# Summary of outliers in each column for customer details file  ; 

Q1 = df1.quantile(0.25)
Q3 = df1.quantile(0.75)
IQR = Q3 - Q1 
lower_limit = Q1 - 1.5*IQR
upper_limit = Q3 + 1.5*IQR
((df1 > upper_limit) | (df1 < lower_limit)).sum()
```

```python
# Summary of outliers in each column for customer policy details file  ; 

Q1 = df2.quantile(0.25)
Q3 = df2.quantile(0.75)
IQR = Q3 - Q1 
lower_limit = Q1 - 1.5*IQR
upper_limit = Q3 + 1.5*IQR
((df2 > upper_limit) | (df2 < lower_limit)).sum()
```

```python
# Summary of outliers  for annual_premium  ; 

Q1 = df2.annual_premium.quantile(0.25)
Q3 = df2.annual_premium.quantile(0.75)
IQR = Q3 - Q1 
LL = Q1 - 1.5*IQR
UL = Q3 + 1.5*IQR
df2.loc[(df2.annual_premium > UL) | (df2.annual_premium < LL)]
print(LL,UL)
```

```python
# Replacing outliers in annual premium which contains a  numeric data 
avg = (df2.loc[(df2['annual_premium'] > LL) & (df2['annual_premium'] < UL) , 'annual_premium' ]).mean()

index1 = df2.loc[(df2['annual_premium'] < LL) | (df2['annual_premium'] > UL) ].index 

print(index1)
df2['annual_premium'] = df2['annual_premium'].replace([df2.loc[index1,'annual_premium']],avg)
```

```python
# Now checking outliers in annual_premium
((df2 > upper_limit) | (df2 < lower_limit)).sum()
```

#### C) Removing White Spaces : 

```python
# Removing white spaces from strings 
df1['Gender'] = df1['Gender'].str.strip()
df1['vehicle_age'] = df1['vehicle_age'].str.strip()
df1['vehicle_damage'] = df1['vehicle_damage'].str.strip()
```

#### F) Drop duplicate Rows : 

```python
df1 = df1.drop_duplicates(keep = False)
df2 = df2.drop_duplicates(keep = False)
```

### 3.Create a Master table for future use. Join the customer table and customer_policy table to get a master table using customer_id in both tables.

```python
master = pd.merge(df1,df2, on = 'customer_id')
master
```
